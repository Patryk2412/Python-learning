```python
# B9. Policz, ilu pracowników ma w tytule słowo „manager” (bez rozróżniania wielkości liter).
sal["is_manager"] = sal["JobTitle"].str.contains("manager", case=False, na=False)
sal["is_manager"].sum()
```

```python
# B10. Dodaj kolumny z udziałami BasePay, OvertimePay, OtherPay w TotalPayBenefits (tylko dla TotalPayBenefits > 0) i opisz rozkład.
pos = sal["TotalPayBenefits"] > 0
sal.loc[pos, "share_base"] = sal.loc[pos, "BasePay"] / sal.loc[pos, "TotalPayBenefits"]
sal.loc[pos, "share_ot"] = sal.loc[pos, "OvertimePay"] / sal.loc[pos, "TotalPayBenefits"]
sal.loc[pos, "share_other"] = sal.loc[pos, "OtherPay"] / sal.loc[pos, "TotalPayBenefits"]
sal.loc[pos, ["share_base","share_ot","share_other"]].describe().T
```

```python
# C. Przekroje po latach
```

```python
# C11. Policz średnie BasePay i TotalPayBenefits dla każdego roku.
sal.groupby("Year")[["BasePay","TotalPayBenefits"]].mean().round(2)
```

```python
# C12. Dla każdego roku wybierz TOP-5 osób według TotalPayBenefits (pokaż rok, imię i nazwisko, stanowisko, TTB).
top5_per_year = (
sal.sort_values(["Year","TotalPayBenefits"], ascending=[True,False])
.groupby("Year").head(5)[["Year","EmployeeName","JobTitle","TotalPayBenefits"]]
)
top5_per_year
```

```python
# C13. Zbuduj tabelę przestawną (index=Year, kolumny=JobTitle_clean, wartości=średnie TTB) i pokaż 10 kolumn o najwyższej średniej w całym okresie.
sal["JobTitle_clean"] = (
sal["JobTitle"].fillna("")
.str.strip().str.upper().str.replace(r"\s+", " ", regex=True)
)


pv = pd.pivot_table(
sal, index="Year", columns="JobTitle_clean",
values="TotalPayBenefits", aggfunc="mean"
)


top_cols = pv.mean().sort_values(ascending=False).head(10).index
pv[top_cols].round(2)
```

```python
# D. Ci sami pracownicy przez lata
```

```python
# D14. Znajdź pracowników, którzy występują co najmniej 2 razy w danych.
name_counts = sal["EmployeeName"].value_counts()
multi_year_names = name_counts[name_counts >= 2].index
len(multi_year_names)
```

```python
# D15. Dla osób z ≥2 latami policz różnicę r/r TotalPayBenefits, a następnie wypisz TOP-10 wzrostów i TOP-10 spadków
chg = (
sal[sal["EmployeeName"].isin(multi_year_names)]
.sort_values(["EmployeeName","Year"])
.groupby("EmployeeName")["TotalPayBenefits"].diff()
.rename("dtpb")
)


sal_chg = pd.concat([sal[["EmployeeName","Year","JobTitle","TotalPayBenefits"]], chg], axis=1)


top_ups = sal_chg.dropna().sort_values("dtpb", ascending=False).head(10)
top_downs = sal_chg.dropna().sort_values("dtpb", ascending=True).head(10)


print(top_ups, top_downs)
```

```python
# D16. Policz, dla ilu osób TotalPayBenefits rósł przez ≥3 kolejne lata.
g = (sal[sal["EmployeeName"].isin(multi_year_names)]
.sort_values(["EmployeeName","Year"])
.groupby("EmployeeName")["TotalPayBenefits"])


def has_3_increase(s):
    s = s.dropna()
    if len(s) < 4:
        return False
    diffs = s.diff().dropna() > 0
    runs = (diffs != diffs.shift()).cumsum().groupby(diffs).transform("size")
    return (diffs & (runs >= 3)).any()

sum(g.apply(has_3_increase))
```

```python
# E17. Dodaj title_len = len(JobTitle) i sprawdź korelację z TotalPayBenefits.
sal["title_len"] = sal["JobTitle"].fillna("").str.len()
sal[["title_len","TotalPayBenefits"]].corr()
```

```python
# E18. Wyekstrahuj słowa z JobTitle (tokeny ≥3 znaki) i podaj 3 najczęstsze w całym zbiorze.
tokens = (
sal["JobTitle"].fillna("")
.str.lower().str.replace(r"[^a-z\s]", " ", regex=True)
.str.split().explode()
)
common_words = tokens[tokens.str.len() >= 3].value_counts().head(20)
common_words.head(3)
```

```python
# E19. Porównaj korelację OvertimePay z TotalPay vs z TotalPayBenefits.
corr_total = sal[["OvertimePay","TotalPay"]].corr().loc["OvertimePay","TotalPay"]
corr_benefit = sal[["OvertimePay","TotalPayBenefits"]].corr().loc["OvertimePay","TotalPayBenefits"]
(corr_total, corr_benefit)
```

```python
# E20. Zdefiniuj wysoki udział nadgodzin (share_ot ≥ 0.30, TTB>0), policz % w każdym roku i pokaż TOP-10 stanowisk w tej grupie.
hi_ot = sal[(sal["TotalPayBenefits"] > 0) & (sal["share_ot"] >= 0.30)]


pct_by_year = (hi_ot.groupby("Year").size() / sal[sal["TotalPayBenefits"] > 0]
.groupby("Year").size() * 100).round(2)


top_jobs_hi_ot = hi_ot["JobTitle_clean"].value_counts().head(10)


print(pct_by_year, top_jobs_hi_ot)
```
