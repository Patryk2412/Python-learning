```python
import numpy as np
```

```python
# 1. Stwórz tablicę 10 zer

arr1 = np.zeros(10)
print("Tablica 10 zer: \n", arr1)
```

```python
# 2. Stwórz tablicę 10 jedynek
arr2 = np.ones(10)
print("Tablica 10 jedynek: \n", arr2)
```

```python
# 3️. Stwórz tablicę 10 piątek

arr_initial = np.ones(10)
arr3 = 5*arr_initial
print("Tablica 10 piątek: \n", arr3)
```

```python
# 4️. Stwórz tablicę z liczbami od 10 do 50

arr4 = np.arange(10,51)
print("Tablica 1-50: \n", arr4)
```

```python
# 5️. Stwórz tablicę z parzystych liczb od 10 do 50
arr5 = np.arange(10,51,2)
print("Tablica liczb parzystych z przedziału [10;50]: \n", arr5)
```

```python
# 6️. Stwórz macierz 3x3 z wartościami od 0 do 8
arr_initial = np.arange(0,9)
arr6 = arr_initial.reshape(3,3)
print("Macierz 3x3 z wartościami 0-8: \n", arr6)
```

```python
# 7. Stwórz macierz jednostkową 5x5
arr7 = np.eye(5)
print("Macierz jednostkowa 5x5: \n", arr7)
```

```python
# 8. Wylosuj 10 liczb z rozkładu jednostajnego [0,1)

arr8 = np.random.rand(10)
print("10 losowych liczb z rozkładu jednostajnego [0,1): \n", arr8)
```

```python
# 9️. Wylosuj macierz 10x10 i znajdź jej wartości min i max

arr9 = np.random.rand(10,10)
print("Macierz losowych liczb z przedziału jednostajnego [0,1): \n", arr9)
print("Minimalna wartość w tej macierzy: ", arr9.min())
print("Maksymalna wartość w tej macierzy: ", arr9.max())
```

```python
# 10. Stwórz tablicę 20 losowych liczb całkowitych 0–100 i znajdź ich średnią

arr10 = np.random.randint(0,101,20)
print("20 losowych liczb z przedziału [0,100]: \n", arr10)
print("Średnia z tych liczb: ", arr10.mean())
```

```python
# 11. 

# Utwórz tablicę arr = np.arange(1, 21).
arr = np.arange(1,21)
print("Stworzona tablica: ", arr)

# Wypisz pierwsze 5 elementów
print("Pierwsze 5 elementów: ", arr[:5])

# Wypisz ostatnie 5 elementów
print("Ostatnie 5 elementów: ", arr[-5:])

# Wypisz tylko liczby parzyste
print("Parzyste elementy tablicy: ", arr[arr %2 == 0])

# Policz średnią, medianę i sumę wszystkich elementów
print("Średnia wszystkich elementów tablicy:  ", np.mean(arr))
print("Mediana wszystkich elementów tablicy:  ", np.median(arr))
print("Suma wszystkich elementów tablicy:  ", np.sum(arr))
```

```python
# 12. Operacje matematyczne

# Dla tej samej tablicy arr:
# Podnieś wszystkie liczby do kwadratu
arr_pow = arr**2
print("Nowa tablica: ", arr_pow)

# Policz pierwiastek kwadratowy każdego elementu
arr_root = np.sqrt(arr)
print("Po spierwiastkowaniu: ", arr_root)

# Oblicz log10 każdego elementu
arr_log = np.log10(arr)
print("Po zlogarytmowaniu: ", arr_log)
```

```python
# 13. Losowe dane i indeksowanie

# Wygeneruj tablicę np.random.randint(0, 100, 15)
arr = np.random.randint(0,100,15) # Zauważ, że nie musisz już tego przekształcać za pomocą array(), bo to juz jest tablica
print("Wylosowana tablica: ", arr)

# Wypisz elementy większe niż 50
print("Elementy tablicy większe od 50: ", arr[arr>50])

# Zamień wszystkie elementy > 80 na 80
arr[arr > 80] = 80
print("Tablica po zmianie: ", arr)


# Znajdź indeks największej wartości (np.argmax)
print("Indeks o największej wartości w tablicy: ", np.argmax(arr))
```

```python
''' 14. Selekcja logiczna

Dla arr = np.arange(1, 16) wybierz liczby: 
większe niż 5, ale mniejsze niż 10,
lub podzielne przez 3.
Połącz oba warunki operatorem logicznym | (OR). '''

arr = np.arange(1,16)
print("Pierwotna tablica: ", arr)
print("Liczby z przedziału (5,10): ", arr[((arr>5) & (arr<10)) | (arr %3 == 0)])
```

```python
# 15. 

#  Utwórz tablicę np.arange(50).reshape(5,10) i wykonaj:
arr = np.arange(50).reshape(5,10)
print("Pierwotna tablica: \n", arr)

# wypisz pierwszy wiersz,
print("Pierwszy wiersz \n", arr[0, :])

# wypisz ostatnią kolumnę,
print("Ostatnia kolumna: \n", arr[:, -1])

# pobierz element z trzeciego wiersza i czwartej kolumny,
print(arr[2][3])

# pobierz submacierz 2x3 z prawego dolnego rogu.
print(arr[3:, 7:])
```

```python
# 16. Statystyka w 2D

print("Pierwotna tablica: \n", arr)

# Dla tej samej tablicy oblicz:
# średnią w każdym wierszu (axis=1),
print("Średnia każdego wiersza w tablicy: ", arr.mean(axis=1))

#maksimum w każdej kolumnie (axis=0),
print("Maksimum z każdej kolumny: ", arr.max(axis=0))

#globalne minimum i maksimum.
print("Maksymalna wartość w tablicy: ", arr.max())
print("Minimalna wartość w tablicy: ", arr.min())
```

```python
# 17. Warunki logiczne

print("Pierwotna tablica: \n", arr)
# Wybierz wszystkie elementy większe niż 30,
print("Elementy tablicy większe od 30: \n", arr[arr>30])

# Stwórz nową macierz zawierającą tylko te elementy,
arr_30 = arr[arr>30]
print("Nowa macierz: \n", arr_30)

# Zamień wszystkie wartości < 10 na 0.
arr_copy = arr.copy()  # Robimy to, żeby nie zmienić pierwotnej tablicy
arr_copy[arr_copy < 10] = 0
print("Po zmianie: \n", arr_copy)
```

```python
# 18. Bonus — losowa macierz

# Wylosuj macierz np.random.randint(1,101, (4,4)),
arr = np.random.randint(1,101, (4,4))
print("Wylosowana macierz: \n", arr)

# Policz sumę w każdej kolumnie i w każdym wierszu,
print("Suma w każdej wierszu: \n", arr.sum(axis=1))
print("Suma w każdym kolumnie: \n", arr.sum(axis=0))


# Znajdź wartość maksymalną i jej współrzędne (wiersz, kolumna).
print("Wartość maksymalna w macierzy: \n", arr.max())

# Znajdź współrzędne (wiersz, kolumna) tej wartości
max_coords = np.unravel_index(np.argmax(arr), arr.shape) # np.argmax(arr) zwraca pozycję w wersji spłaszczonej (1D).
                                                         # np.unravel_index(..., arr.shape) przelicza ją na indeksy (wiersz, kolumna) w macierzy 2D.
print("Współrzędne maksymalnej wartości (wiersz, kolumna):", max_coords)
```
