```python
''' 1. Division Calculator
Napisz prosty kalkulator, który:
pyta użytkownika o dwie liczby,
dzieli pierwszą przez drugą,
łapie ZeroDivisionError i ValueError,
w bloku finally wypisuje "Koniec programu. ''' 

try: 
    number1 = int(input("Podaj pierwszą liczbę: "))
    number2 = int(input("Podaj drugą liczbę: "))
    
    result = number1/number2
    
except ZeroDivisionError:
    print("Nie można dzieloć przez 0!")
    
except ValueError: 
    print("Podano błędną wartość!") 

except Exception as e: 
    print("Inny błąd: ", e) 

else: 
    print("Wynik dzielenia: ") 
    print(result) 

finally: 
    print("Koniec programu.") 
```

```python
''' Zadanie 2 – File Reader
Napisz program, który:
Pyta użytkownika o nazwę pliku.
Próbuje otworzyć ten plik i policzyć, ile jest w nim linii.
Jeśli pliku nie ma → łapie FileNotFoundError i wypisuje komunikat.
Jeśli plik uda się otworzyć → wypisuje liczbę linii w bloku else.
W finally wypisuje "Koniec programu." '''

name = input("Podaj nazwę pliku: ") 

try:
    file = open(name, "r")
    lines = file.readlines()   # readlines() zamienia plik na listę stringów, gdzie każdy element listy = jedna linia z pliku.
    
except FileNotFoundError: 
    print("Nie ma pliku o takiej nazwie. ") 

else: 
    print("Otwarty plik ma {} linii. ".format(len(lines)))
    file.close()   # Pamiętaj o zamknięciu pliku!
          
finally:
    print("Koniec programu. ")
```

```python
''' Zadanie 3 – Safe Input

Napisz program, który:

Pyta użytkownika o liczbę całkowitą.
Jeśli poda coś innego → łapie ValueError i prosi jeszcze raz.
Gdy użytkownik poda poprawną liczbę → wypisuje ją i kończy działanie. '''

while True:   # nieskończona pętla
    try:
            liczba = int(input("Podaj liczbę całkowitą: "))

    except ValueError:
        print("Podałeś błędną wartość.") 

    else: 
        print("Dziękuję. Podana liczba: ", liczba) 
        break  # jak wartość jest ok, to przerywamy program

print("Koniec programu.") 
```

```python
''' Zadanie 4 – Dictionary Lookup

Napisz program, który:

Tworzy prosty słownik, np. {"PL": "Polska", "DE": "Niemcy", "FR": "Francja"}.
Pyta użytkownika o kod kraju.
Próbuje znaleźć kraj w słowniku.
Jeśli nie znajdzie → łapie KeyError i wypisuje komunikat.
Jeśli znajdzie → wypisuje nazwę kraju. I pyta aż do skutku. 
'''

dic = {"PL": "Polska", "DE": "Niemcy", "FR": "Francja"}

while True: 
    try: 
        kod = input("Podaj kod kraju: ")
        kraj = dic[kod]

    except KeyError:
        print("Błędny kod. Podaj poprawny. ")

    else: 
        print("Nazwa kraju o podanym kodzie: ", kraj)
        break
        
    finally: 
        print("Próba zakończona. ") 
```

```python
''' Zadanie 5 – List Index

Napisz program, który:

Ma listę np. nums = [10, 20, 30].
Pyta użytkownika o indeks.
Próbuj wypisać element o tym indeksie.
Jeśli indeks jest poza zakresem → łapie IndexError i wypisuje komunikat.
Jeśli użytkownik poda coś złego (np. litery zamiast liczby) → łapie ValueError.
W else pokazuje element listy.
W finally wypisuje "Koniec programu.". '''

nums = [10,20,30] 

try: 
    indeks = int(input("Podaj indeks: "))
    print(nums[indeks])
    
except IndexError:
    print("Podałeś indeks będący poza zakresem. ")

except ValueError: 
    print("Podałeś błędną wartość.") 

finally: 
    print("Koniec programu.") 
    
```
